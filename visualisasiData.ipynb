```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Membuat DataFrame
data = {
    'Minggu': [1, 2, 3, 4, 5],
    'Produk A': [50, 60, 70, 80, 90],
    'Produk B': [30, 25, 35, 40, 50],
    'Produk C': [20, 30, 25, 35, 45]
}
df = pd.DataFrame(data)

# Line Chart
plt.figure(figsize=(10, 5))
plt.plot(df['Minggu'], df['Produk A'], marker='o', label='Produk A')
plt.plot(df['Minggu'], df['Produk B'], marker='s', label='Produk B')
plt.plot(df['Minggu'], df['Produk C'], marker='^', label='Produk C')
plt.title('Penjualan Produk Per Minggu')
plt.xlabel('Minggu')
plt.ylabel('Jumlah Terjual')
plt.legend()
plt.grid(True)
plt.show()

# Bar Chart
df.set_index('Minggu').plot(kind='bar', figsize=(10, 5), title='Penjualan Produk Per Minggu')
plt.xlabel('Minggu')
plt.ylabel('Jumlah Terjual')
plt.show()

# Pie Chart
plt.figure(figsize=(7, 7))
plt.pie(df.set_index('Minggu').sum(), labels=['Produk A', 'Produk B', 'Produk C'], autopct='%1.1f%%', startangle=140)
plt.title('Proporsi Total Penjualan Produk')
plt.show()

# Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df.drop(columns='Minggu'))
plt.title('Distribusi Penjualan Produk')
plt.show()

# Scatter Plot
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['Minggu'], y=df['Produk A'], label='Produk A', s=100)
sns.scatterplot(x=df['Minggu'], y=df['Produk B'], label='Produk B', s=100)
sns.scatterplot(x=df['Minggu'], y=df['Produk C'], label='Produk C', s=100)
plt.title('Scatter Plot Penjualan Produk')
plt.xlabel('Minggu')
plt.ylabel('Jumlah Terjual')
plt.legend()
plt.show()
```
